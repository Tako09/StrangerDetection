#フェイスディテクション撮影  
下記の処理を行うプログラム  
・カメラの立ち上げ  
・顔の認識  
・認識後撮影  
・カメラを閉じる

```python
# ライブラリのインストール
import cv2 as cv
import os
from time import time
from io import BytesIO
import IPython
import datetime
# os.environ["OPENCV_VIDEOIO_MSMF_ENABLE_HW_TRANSFORMS"] = "0"
```

```python
# 変数の定義
cascadePath = 'cascade/lbpcascade_frontalface_improved.xml'
assert os.path.isfile(cascadePath), 'lbpcascade_frontalface_improved.xml がない'
# 分類気の設定
frontalFaceCascadeClassifier = cv.CascadeClassifier(cascadePath)
```

```python
def play_demo(camera=0, play_time=5):
    # カメラを立ち上げて一定時間表示する関数
    # 顔が出ている場合検知を知らせる
    
    # IPythonディスプレイ準備
    lab_display = IPython.display.display("", display_id=1)

    # カメラキャプチャ設定
    cap = cv.VideoCapture(camera)
    cap.set(cv.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv.CAP_PROP_FRAME_HEIGHT, 360)
    
    while cap.isOpened():
        # カメラ画像取得
        ret, frame = cap.read() # 画像データの取得
        returnFrame = frame
        # cv.imshow("Frame", frame) # ウィンドウに出力
        
        grayImage = cv.cvtColor(frame, cv.COLOR_BGR2GRAY) # グレースケールに変更
        grayImage = cv.equalizeHist(grayImage) # データを平たん化する
        multipleFaces = frontalFaceCascadeClassifier.detectMultiScale(grayImage, 1.1, 3, 0, (20, 20)) # 顔検出
        lineColor = (255, 102, 51) # 表示数る線の色
        fontColor = (255, 102, 51) # 文字の色
        # for (x, y, width, height) in multipleFaces:
        #     # 見つかった顔を線で囲む
        #     cv.rectangle(frame, (x, y), (x + width, y + height), lineColor, 2)
        #     cv.putText(frame, 'Face Dtected', (x, y), cv.FONT_HERSHEY_SIMPLEX, 0.7, fontColor, 1, cv.LINE_AA)
        # ディスプレイ表示
        _, frame = cv.imencode('.jpg', frame, [int(cv.IMWRITE_JPEG_QUALITY), 100])
        frame_buffer = BytesIO(frame)
        lab_display.update(IPython.display.Image(data=frame_buffer.getvalue()))
        
        if len(multipleFaces) <= 0:
            continue
        else:
            print('終了します')
            cap.release()
            cv.destroyAllWindows()
            return returnFrame, multipleFaces # ぶれてない写真を保存して返したい
```

```python
def show_img(img):
    from IPython.display import Image, display
    # インラインで画像を表示する
    img = cv.imencode('.png', img)[1]
    display(Image(img))
```

```python
def save_img(frame):
    # ファイルを保存する
    now = str(datetime.datetime.now())[0:19].replace(' ', '_')
    path = 'data/' + now + '.png'
    cv.imwrite(path, frame)
    print('保存完了')
```

```python
nowName = str(datetime.datetime.now())[0:19].replace(' ', '_')
path = nowName + '.png'
ret = cv.imwrite(path, frame)
assert ret, "出力に失敗"
```

```python
# カメラの起動が遅い・・・。
# rasberyでも遅かったら考える。
frame, multipleFaces = play_demo(camera=0, play_time=30)
# print(frame)
```

```python
show_img(frame) # 表示
save_img(frame) # 保存
```
